```python
import pandas as pd
```

```python
df=pd.read_csv('Airlines.csv')
```

```python
df = df.sample(frac=0.2, random_state=42)  # فقط 20% از داده‌ها
```

```python
df['Delay'].value_counts(normalize=True)
```

```python
df.info()
```

```python
Dummy_1 = pd.get_dummies(df[['Airline', 'AirportFrom', 'AirportTo']], dtype=int)
```

```python
df=df.drop(['Airline','AirportFrom','AirportTo'],axis=1)
```

```python
df=pd.concat([df,Dummy_1],axis=1)
```

```python
df.head(5)
```

```python
df.info()
```

```python
X=df.drop('Delay',axis=1)
Y=df['Delay']
```

Standardizing

```python
from sklearn import preprocessing
transform=preprocessing.StandardScaler()
X_scaled=transform.fit_transform(X)
col=X.columns
X=pd.DataFrame(X_scaled,columns=col)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=101)
from sklearn.linear_model import LogisticRegression
airmodel=LogisticRegression()
airmodel.fit(X_train,y_train)
```

```python
y_train.value_counts(normalize=True)
```

```python
predictions=airmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

### KNeighborClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train,y_train)
```

```python
predictions_knn=knn.predict(X_test)
```

```python
confusion_matrix(y_test,predictions_knn)
```

```python
print(classification_report(y_test,predictions_knn))
```

Support Vector Machine

```python
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm = LinearSVC()
svm.fit(X_train_scaled, y_train)
predictions = svm.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# دقت کلی
print("Accuracy:", accuracy_score(y_test, predictions))

# گزارش کامل شامل precision، recall، f1
print("\nClassification Report:\n", classification_report(y_test, predictions))

# ماتریس درهم‌ریختگی (confusion matrix)
print("\nConfusion Matrix:\n", confusion_matrix(y_test, predictions))
```

DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(X_train,y_train)
predictions_tree=tree.predict(X_test)
confusion_matrix(y_test,predictions_tree)
print(classification_report(y_test,predictions_tree))
```
